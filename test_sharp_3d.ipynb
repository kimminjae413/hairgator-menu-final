```python
# 3. 테스트 이미지 다운로드 (FAL0001 스타일)
import urllib.request
import os

os.makedirs('input', exist_ok=True)
os.makedirs('output', exist_ok=True)

# Firebase Storage에서 헤어 스타일 이미지 다운로드
image_url = "https://firebasestorage.googleapis.com/v0/b/hairgator-70a28.firebasestorage.app/o/styles%2FFAL0001%2Fresult.png?alt=media"

urllib.request.urlretrieve(image_url, 'input/FAL0001.png')
print('이미지 다운로드 완료!')

# 이미지 확인
from PIL import Image
from IPython.display import display
img = Image.open('input/FAL0001.png')
display(img)
```

```python
# 4. SHARP로 3D 변환
!sharp predict -i input -o output
print('\n3D 변환 완료!')
```

```python
# 5. 결과 파일 확인
import os
for f in os.listdir('output'):
    path = os.path.join('output', f)
    size_mb = os.path.getsize(path) / 1024 / 1024
    print(f'{f}: {size_mb:.2f} MB')
```
